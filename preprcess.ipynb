```python
import pandas as pd
import numpy as np
```

```python
# read files
require = pd.read_csv("Requirements.csv",header=None,names =['Rid', 'title', 'content'], error_bad_lines=False)
achieve = pd.read_csv("Train_Achievements.csv",header=None,names =['Aid', 'title', 'content'],index_col=False,error_bad_lines=False)
relation = pd.read_csv("Train_Interrelation.csv", error_bad_lines=False)
```

```python
a = require['Rid'].tolist()
a = [i[1:-2] for i in a]
a=pd.Series(a)
require['Rid']=a

b = achieve['Aid'].tolist()
b = [i[1:-2] for i in b]
b=pd.Series(b)
achieve['Aid']=b
```

```python
mid1 = relation.merge(achieve, on='Aid')
```

```python
mid2 = mid1.merge(require,on='Rid')
```
